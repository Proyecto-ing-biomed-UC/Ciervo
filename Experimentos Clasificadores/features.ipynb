```python
data  = example_marcha()[0] 

print(data.head())

# spectrogram
transform = torchaudio.transforms.Spectrogram(n_fft=32)
spectrogram = transform(torch.tensor(data['AductorLargo'].tolist()))


# plot spectrogram
spec_db = torchaudio.transforms.AmplitudeToDB('power',top_db=40)(spectrogram)

plt.figure(figsize=(10, 5))
plt.imshow(spec_db.numpy(), aspect='auto', origin='lower')
plt.colorbar()
plt.show()

```
